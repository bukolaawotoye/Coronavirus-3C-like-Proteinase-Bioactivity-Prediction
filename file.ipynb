```python
#chembl dATatabase
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

Target search virus for coronavirus

```python
#target search for corona virus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#selecting and retrieving bioactivity data for sars coronavirus 3c-like protinase
selected_target = targets.target_chembl_id[6]
selected_target
```

```python
#retrieving  only bioactivity data for corona virus 3c-like proteinase (CHEMBL3927) that are reported as IC50 values in nM (nanomolar) unit
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
df.head()
```

The standard value column in the dataframe indicates the potency of the drugs, the lower the number the better the potency of the drug becomes
the lower the concentration of the drug the better it is, because it is better to require fewer drugs to create inhibition at 50%
IC50 - inhibtor concentration at 50%

```python
df.standard_type.unique()
```

```python
#saving the biosctivity data to csv
df.to_csv('bioactivity_data.csv', index=False)
```

```python
data = pd.read_csv('bioactivity_data.csv')
data.head(5)
```

```python
#handling missing values 
data = data[data.standard_value.notna()]
data
```

DATA PREPROCESSING
Compounds having values less than 1000nM will be considered active, whie those greater than 10,000nM will be considered inactive, as for those betweeen 1000, and 10,000nM will be referred to as intermediate

```python
bioactivity_class = []
for i in data.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append('inactive')
    elif float(i) <= 1000:
        bioactivity_class.append('active')
    else:
        bioactivity_class.append('intermediate')
```

```python
#molecule chamber id: the identification number of each molecule
#molecule: chemical structure, eexerts effects on the target protein in order to produce the desired bioligical effects 
mol_cid = []
for i in data.molecule_chembl_id:
    mol_cid.append(i)
```

```python
#iterate canonical_smiles to a list
canonical_smiles = []
for i in data.canonical_smiles:
    canonical_smiles.append(i)
```

```python
#iterate standard_value to a list
standard_value = []
for i in data.standard_value:
    standard_value.append(i)
```

```python
#combining the 4 lists into a dataframe
data_tuples = list(zip(mol_cid,canonical_smiles, bioactivity_class,standard_value))
df3 = pd.DataFrame(data_tuples, columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```
